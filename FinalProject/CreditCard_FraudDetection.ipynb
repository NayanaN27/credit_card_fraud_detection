```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, average_precision_score
import numpy as np
```

```python
classes = np.unique(y_train)

class_weights_array = compute_class_weight(
    class_weight="balanced",
    classes=classes,
    y=y_train
)

class_weights = {
    0: class_weights_array[0],
    1: class_weights_array[1]
}

class_weights
```

Build improved neural network

```python
improved_nn = Sequential([
    Dense(64, activation="relu", input_shape=(X_train_scaled.shape[1],)),
    BatchNormalization(),
    Dropout(0.3),

    Dense(32, activation="relu"),
    BatchNormalization(),
    Dropout(0.3),

    Dense(16, activation="relu"),
    Dropout(0.2),

    Dense(1, activation="sigmoid")
])

improved_nn.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="binary_crossentropy",
    metrics=[
        "accuracy",
        tf.keras.metrics.AUC(name="roc_auc"),
        tf.keras.metrics.AUC(name="pr_auc", curve="PR")
    ]
)

improved_nn.summary()
```

Train improved NN

```python
early_stop = EarlyStopping(
    monitor="val_pr_auc",
    mode="max",
    patience=5,
    restore_best_weights=True
)

reduce_lr = ReduceLROnPlateau(
    monitor="val_pr_auc",
    mode="max",
    factor=0.5,
    patience=3,
    min_lr=1e-6
)

history_improved_nn = improved_nn.fit(
    X_train_scaled,
    y_train,
    validation_data=(X_val_scaled, y_val),
    epochs=50,
    batch_size=2048,
    class_weight=class_weights,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)
```

Evaluate improved NN at threshold 0.5

```python
improved_nn_proba = improved_nn.predict(X_test_scaled).ravel()

improved_nn_pred_05 = (improved_nn_proba >= 0.5).astype(int)

print("Improved Neural Network Classification Report:")
print(classification_report(y_test, improved_nn_pred_05))

print("Improved Neural Network Confusion Matrix:")
print(confusion_matrix(y_test, improved_nn_pred_05))

print("Precision:", precision_score(y_test, improved_nn_pred_05))
print("Recall:", recall_score(y_test, improved_nn_pred_05))
print("F1-score:", f1_score(y_test, improved_nn_pred_05))
print("ROC-AUC:", roc_auc_score(y_test, improved_nn_proba))
print("PR-AUC:", average_precision_score(y_test, improved_nn_proba))
```

Tune NN threshold

```python
val_improved_nn_proba = improved_nn.predict(X_val_scaled).ravel()

thresholds = np.arange(0.05, 0.96, 0.01)

nn_threshold_results = []

for threshold in thresholds:
    y_pred_threshold = (val_improved_nn_proba >= threshold).astype(int)

    precision = precision_score(y_val, y_pred_threshold, zero_division=0)
    recall = recall_score(y_val, y_pred_threshold, zero_division=0)
    f1 = f1_score(y_val, y_pred_threshold, zero_division=0)

    nn_threshold_results.append({
        "threshold": threshold,
        "precision": precision,
        "recall": recall,
        "f1_score": f1
    })

nn_threshold_df = pd.DataFrame(nn_threshold_results)

nn_threshold_df.sort_values(by="f1_score", ascending=False).head(10)
```

```python
best_nn_threshold = nn_threshold_df.sort_values(
    by="f1_score",
    ascending=False
).iloc[0]["threshold"]

print("Best NN threshold from validation set:", best_nn_threshold)

improved_nn_test_proba = improved_nn.predict(X_test_scaled).ravel()
improved_nn_best_pred = (improved_nn_test_proba >= best_nn_threshold).astype(int)

print("Improved Neural Network Test Results")
print(confusion_matrix(y_test, improved_nn_best_pred))
print(classification_report(y_test, improved_nn_best_pred))

improved_nn_precision = precision_score(y_test, improved_nn_best_pred, zero_division=0)
improved_nn_recall = recall_score(y_test, improved_nn_best_pred, zero_division=0)
improved_nn_f1 = f1_score(y_test, improved_nn_best_pred, zero_division=0)
improved_nn_roc_auc = roc_auc_score(y_test, improved_nn_test_proba)
improved_nn_pr_auc = average_precision_score(y_test, improved_nn_test_proba)

print("Precision:", improved_nn_precision)
print("Recall:", improved_nn_recall)
print("F1-score:", improved_nn_f1)
print("ROC-AUC:", improved_nn_roc_auc)
print("PR-AUC:", improved_nn_pr_auc)
```

## Neural Network Architecture Comparison

To strengthen the neural network part of the project, we compare shallow, deep, and dropout-based neural network architectures. This helps us analyze whether increasing model depth or adding regularization improves fraud detection performance on imbalanced transaction data.

Use best NN threshold

```python
def build_nn_model(model_type="shallow"):
    model = keras.Sequential()
    model.add(keras.Input(shape=(X_train_scaled.shape[1],)))

    if model_type == "shallow":
        model.add(layers.Dense(32, activation="relu"))

    elif model_type == "deep":
        model.add(layers.Dense(64, activation="relu"))
        model.add(layers.Dense(32, activation="relu"))
        model.add(layers.Dense(16, activation="relu"))

    elif model_type == "dropout":
        model.add(layers.Dense(64, activation="relu"))
        model.add(layers.Dropout(0.3))
        model.add(layers.Dense(32, activation="relu"))
        model.add(layers.Dropout(0.3))
        model.add(layers.Dense(16, activation="relu"))

    model.add(layers.Dense(1, activation="sigmoid"))

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss="binary_crossentropy",
        metrics=[
            "accuracy",
            keras.metrics.AUC(name="roc_auc"),
            keras.metrics.AUC(name="pr_auc", curve="PR")
        ]
    )

    return model
```

```python
nn_models = {
    "Shallow NN": build_nn_model("shallow"),
    "Deep NN": build_nn_model("deep"),
    "Deep NN + Dropout": build_nn_model("dropout")
}

nn_results = []

for name, model in nn_models.items():
    print("\nTraining:", name)

    model.fit(
        X_train_scaled,
        y_train,
        validation_data=(X_val_scaled, y_val),
        epochs=20,
        batch_size=2048,
        class_weight=class_weights,
        callbacks=[early_stop],
        verbose=1
    )

    val_proba = model.predict(X_val_scaled).ravel()

    thresholds = np.arange(0.05, 0.96, 0.01)
    best_f1 = 0
    best_threshold = 0.5

    for threshold in thresholds:
        val_pred = (val_proba >= threshold).astype(int)
        f1 = f1_score(y_val, val_pred, zero_division=0)

        if f1 > best_f1:
            best_f1 = f1
            best_threshold = threshold

    test_proba = model.predict(X_test_scaled).ravel()
    test_pred = (test_proba >= best_threshold).astype(int)

    nn_results.append({
        "Model": name,
        "Threshold": best_threshold,
        "Precision": precision_score(y_test, test_pred, zero_division=0),
        "Recall": recall_score(y_test, test_pred, zero_division=0),
        "F1-score": f1_score(y_test, test_pred, zero_division=0),
        "ROC-AUC": roc_auc_score(y_test, test_proba),
        "PR-AUC": average_precision_score(y_test, test_proba)
    })
```

```python
nn_comparison_df = pd.DataFrame(nn_results)
nn_comparison_df.sort_values(by="F1-score", ascending=False)
```

```python
best_nn_threshold = nn_threshold_df.sort_values(
    by="f1_score",
    ascending=False
).iloc[0]["threshold"]

print("Best NN threshold:", best_nn_threshold)

improved_nn_pred_best = (improved_nn_proba >= best_nn_threshold).astype(int)

print("Improved NN Classification Report with Best Threshold:")
print(classification_report(y_test, improved_nn_pred_best))

print("Improved NN Confusion Matrix with Best Threshold:")
print(confusion_matrix(y_test, improved_nn_pred_best))

improved_nn_precision = precision_score(y_test, improved_nn_pred_best)
improved_nn_recall = recall_score(y_test, improved_nn_pred_best)
improved_nn_f1 = f1_score(y_test, improved_nn_pred_best)
improved_nn_roc_auc = roc_auc_score(y_test, improved_nn_proba)
improved_nn_pr_auc = average_precision_score(y_test, improved_nn_proba)

print("Improved NN Precision:", improved_nn_precision)
print("Improved NN Recall:", improved_nn_recall)
print("Improved NN F1-score:", improved_nn_f1)
print("Improved NN ROC-AUC:", improved_nn_roc_auc)
print("Improved NN PR-AUC:", improved_nn_pr_auc)
```

### Final Model Selection

Based on the model comparison results, the Random Forest model with threshold 0.3 achieved the highest F1-score. This means it provided the best balance between precision and recall among all tested models.

Therefore, the final selected model for the demo application is Random Forest with threshold 0.3.
